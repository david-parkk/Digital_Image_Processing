```python
import numpy as np
import cv2 as cv
img = cv.imread('karoro.png')
assert img is not None, "file could not be read, check with os.path.exists()"
import warnings
warnings.filterwarnings('ignore')
```

```python
dimension1,dimension2,dimension3=img.shape
```

```python
img.shape
```

```python
img.dtype
```

```python
img=img.astype(np.float64)
```

```python
r=np.zeros((dimension1,dimension2),dtype=np.float64)
g=np.zeros((dimension1,dimension2),dtype=np.float64)
b=np.zeros((dimension1,dimension2),dtype=np.float64)
```

```python
r.dtype
```

```python
img.dtype
```

```python
R=img[:,:,2]
G=img[:,:,1]
B=img[:,:,0]
```

```python
r=img[:,:,2]/(img[:,:,0]+img[:,:,1]+img[:,:,2])
g=img[:,:,1]/(img[:,:,0]+img[:,:,1]+img[:,:,2])
b=img[:,:,0]/(img[:,:,0]+img[:,:,1]+img[:,:,2])
```

```python
r.dtype
```

```python
s=np.zeros((dimension1,dimension2),dtype=np.float64)
```

```python
s.shape
```

```python
s.shape
```

## S구하기

```python
for i in range(dimension1):
    for j in range(dimension2):
        s[i][j]=(1-3*min([r[i][j],g[i][j],b[i][j]]))
    
```

```python
s
```

```python
s=s*255
```

```python
s
```

## I 구하기

```python
inten=np.zeros((dimension1,dimension2),dtype=np.float64)
```

```python
for i in range(dimension1):
    for j in range(dimension2):
        inten[i][j]=(R[i][j]+G[i][j]+B[i][j])/(3)
    
```

```python
inten
```

## H구하기

```python
h=np.zeros((dimension1,dimension2),dtype=np.uint8)
```

```python
for i in range(dimension1):
    for j in range(dimension2):
        if R[i][j]<= B[i][j]:
            temp1=(0.5*(R[i][j]-G[i][j]+R[i][j]-B[i][j]))
            temp2=np.sqrt((R[i][j]-G[i][j])*(R[i][j]-G[i][j])+(R[i][j]-B[i][j])*(G[i][j]-B[i][j]))
            h[i][j]=144*np.arccos((temp1/temp2))
            
            #print(h[i][j])
        else:
            temp1=(0.5*(r[i][j]-g[i][j]+r[i][j]-b[i][j]))
            temp2=np.sqrt((r[i][j]-g[i][j])*(r[i][j]-g[i][j])+(r[i][j]-b[i][j])*(g[i][j]-b[i][j]))
            h[i][j]=144*(2*np.pi-np.arccos((temp1/temp2)))
            
    
```
